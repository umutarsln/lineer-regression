```python
features = []
for season in seasons:    
    driver = webdriver.Chrome()
    driver.get(season)
    soup = BeautifulSoup(driver.page_source, "html.parser")
    m = soup.find("meta" , {"name": "Description"}).get("content").split(" ")[5]
    for i in range(0, len(soup.find("table", {"id": f"results{m}91_overall"}).findAll("td",{"data-stat":"team"}))):   
        try:
            season_ = soup.find("meta" , {"name": "Description"}).get("content").split(" ")[5]
        except:
            season_ = np.nan
        try:
            squad = soup.find("table", {"id": f"results{m}91_overall"}).findAll("td",{"data-stat":"team"})[i].find("a").text.strip()
        except:
            squad = np.nan
        try:
            mp = soup.find("table", {"id": f"results{m}91_overall"}).findAll("td",{"data-stat":"games"})[i].text.strip()
        except:
            mp = np.nan
        try:
            w = soup.find("table", {"id": f"results{m}91_overall"}).findAll("td",{"data-stat":"wins"})[i].text.strip()
        except:
            w = np.nan
        try:
            d = soup.find("table", {"id": f"results{m}91_overall"}).findAll("td",{"data-stat":"ties"})[i].text.strip()
        except:
            d = np.nan
        try:
            l = soup.find("table", {"id": f"results{m}91_overall"}).findAll("td",{"data-stat":"losses"})[i].text.strip()
        except:
            l = np.nan
        try:
            gf = soup.find("table", {"id": f"results{m}91_overall"}).findAll("td",{"data-stat":"goals_for"})[i].text.strip()
        except:
            gf = np.nan
        try:
            ga = soup.find("table", {"id": f"results{m}91_overall"}).findAll("td",{"data-stat":"goals_against"})[i].text.strip()
        except:
            ga = np.nan
        try:
            gd = soup.find("table", {"id": f"results{m}91_overall"}).findAll("td",{"data-stat":"goal_diff"})[i].text.strip()
        except:
            gd = np.nan
        try:
            pts = soup.find("table", {"id": f"results{m}91_overall"}).findAll("td",{"data-stat":"points"})[i].text.strip()
        except:
            pts = np.nan
        try:
            pts_mp = soup.find("table", {"id": f"results{m}91_overall"}).findAll("td",{"data-stat":"points_avg"})[i].text.strip()
        except:
            pts_mp = np.nan
                
        features.append([season_, squad, mp, w, d, l, gf, ga, gd, pts, pts_mp])
        time.sleep(1)
    
    df_columns = ["Sezon","Takım","Toplam Maç","Zafer","Berabere","Yenilgi","GF","GA","GD","Pts","Pts/MP"]

    df = pd.DataFrame.from_records(features, columns=df_columns)
driver.quit()
```
