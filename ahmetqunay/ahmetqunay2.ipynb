```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
import time, os
```

```python
fbref_query = "https://fbref.com/en/comps/9/history/Premier-League-Seasons"
```

```python
def getAndParseURL(url):
    result = requests.get(url, headers={"User-Agent":"Mozilla/5.0"})
    soup = BeautifulSoup(result.text, "html.parser")
    return soup
```

```python
html = getAndParseURL(fbref_query)
```

```python
seasons = []

season_count=74
for season in html.findAll("td",{"data-stat":"league_name"}):
    seasons.append("https://fbref.com" + season.find("a").get("href"))
    season_count -= 1
    if season_count == 0:
        break
        
seasons
```

```python
season_years = []
for season in seasons:
    html = getAndParseURL(season)
    m = html.find("meta", {"name": "Description"}).get("content").split(" ")[5]
    for i in range(0, len(html.find("table", {"id": f"results{m}91_overall"}).findAll("td",{"data-stat":"team"}))):
        try:
            
            current_df = html.find("meta", {"name": "Description"}).get("content").split(" ")[5]
            season_years.append(current_df)
        except:
            print(season)
    time.sleep(5)    
```
