```python
import requests
from bs4 import BeautifulSoup as bts
import pandas as pd
import re
import numpy as np
from datetime import datetime
import time
```

```python
def getAndParseURL(url):
    result = requests.get(url, headers={"User-Agent":"Mozilla/5.0"})
    soup = bts(result.text, "html.parser")
    return soup
```

```python
features = []
b_last_28k = b[-27000:]
for product in b_last_28k:
    html = getAndParseURL(product)
    print(f"{b_last_28k.index(product)}/{len(b_last_28k)}")
    try:
        book_name = html.find("h1",{"class":"pr_header__heading"}).text.strip()
    except:
        book_name = np.nan
    try:
        book_genre = html.find("div", {"class":"breadcrumb"}).findAll("a")[2].text.strip()
    except:
        book_genre = np.nan
    try:
        author_name = html.find("div", {"class":"pr_producers__manufacturer"}).text.strip()
    except:
        author_name = np.nan
    try:
        publisher = html.find("div",{"class":"pr_producers__publisher"}).text.strip()
    except:
        publisher = np.nan   
    try:
        publish_date = html.find("div",{"class":"attributes"}).find(text=re.compile("Yayın Tarihi")).findNext().text.strip().replace(".","/")
    except:
        publish_date = np.nan
    try:
        purchase_info = html.find("div",{"class":"purchase-info"}).text.strip().replace(".","")
    except:       
        purchase_info = np.nan
    try:
        page_number = int(html.find("div",{"class":"attributes"}).find(text=re.compile("Sayfa Sayısı:")).findNext().text.strip())
    except:
        page_number = np.nan
    try:
        cover_type =  html.find("div",{"class":"attributes"}).find(text=re.compile("Cilt Tipi:")).findNext().text.strip()
    except:
        cover_type = np.nan
    try:
        paper_type = html.find("div",{"class":"attributes"}).find(text=re.compile("Kağıt Cinsi:")).findNext().text.strip()
    except:
        paper_type = np.nan
    try:
        size = html.find("div",{"class":"attributes"}).find(text=re.compile("Boyut:")).findNext().text.strip()
    except:
        size = np.nan
    try:
        language = html.find("div",{"class":"attributes"}).find(text=re.compile("Dil:")).findNext().text.strip()
    except:
        language = np.nan
    try:
        rating = html.find("meta", itemprop="ratingValue").attrs["content"].replace(".",",")
    except:
        rating = np.nan
    try:
        rating_count = html.find("meta", itemprop="ratingCount").attrs["content"].replace(".",",")
    except:
        rating_count = np.nan
    try:
        review_count = html.find("meta", itemprop="reviewCount").attrs["content"].replace(".","")
    except:
        review_count = np.nan
    try:
        fav_count = int(html.find("span",{"id":"favorite-count"}).text.strip().replace(".",""))
    except:
        fav_count = np.nan
    try:
        to_read_list = int(html.find("span",{"class":"readlist-count"}).text.strip())
    except:
        to_read_list = np.nan
    try:
        price = float(html.find("div",{"class":"price__item"}).text.strip().replace(",","."))
    except:
        price = np.nan
    try:
        discount_point = html.find("p",{"class":"info-text"}).text.replace(",",".")
    except:
        discount_point = np.nan



    features.append([book_name, book_genre, author_name, publisher, publish_date, purchase_info, page_number, cover_type, paper_type, size, 
             language, rating, rating_count, review_count, fav_count, to_read_list, price, discount_point])
        

df_columns = ["Book Name", "Book Genre","Author Name","Publisher","Publish Date","Purchase Info","Page Numbers","Cover Type",
                  "Paper Type", "Size", "Language","Rate","Rating Count","Review Count","Fav Count","To-Read List Count","Price","Discount"]

df = pd.DataFrame.from_records(features, columns=df_columns)
df.to_csv("./kitap_yurdu_data.csv") 
    
```
