```python
def split_int(string):
    for i in string.split():
        try:
            str_to_num = int(i)
            return str_to_num
        except ValueError:
            pass
df["Purchase Info"] = df.loc[df["Purchase Info"].notnull(), "Purchase Info"].apply(split_int)
```

```python
paper_type_mapping = { 'Kitap Kağıdı':1, '1. Hm. Kağıt':2, '3. Hm. Kağıt':3, 'Kuşe Kağıt':4, 'Şamuha Kağıt':5, '2. Hm. Kağıt':6}

df["Paper Type"] = df["Paper Type"].map(paper_type_mapping)
df["Paper Type"]
```

```python
def split_float(string):
    for i in string.split():
        str = string.split()
        try:
            return float(str[0])*float(str[2])
        except ValueError:
            pass
```

```python
split_float(df.loc[1,"Size"])
```

```python
df["Size"] = df.loc[df["Size"].notnull(), "Size"].apply(split_float)
```

```python
import datetime
import os
```

```python
today = datetime.datetime.now().date()
def time_passed(publish_date):
    rd = datetime.datetime.strptime(publish_date, "%d/%m/%Y").date()
    time_since_publ = (today-rd).days
    return(time_since_publ)
```

```python
df['Time Since Publication'] = df.loc[df["Publish Date"].notnull(), "Publish Date"].apply(time_passed)
```

```python
language_counts = df['Language'].value_counts()

# Identify languages with counts below 150
infrequent_languages = language_counts[language_counts < 150].index

# Create a new category 'Other' for infrequent languages
df['Language_categorized'] = df['Language'].apply(lambda x: 'Other' if x in infrequent_languages else x)
```
